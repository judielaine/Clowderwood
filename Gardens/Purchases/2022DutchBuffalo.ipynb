```python
# Scrape table
localFile = "2022DutchBuffalo.pdf"
tables = camelot.read_pdf(localFile,pages='1,2,3')
tables
```

```python
# Clean up table
df = pandas.concat( [tables[0].df,tables[1].df,tables[2].df])
colList = list(df.iloc[0])
colList[2] = "Origin code"
df.columns = colList
df = df.drop(0)
```

```python
# Add fields for perennial airtable
df["Name"] = df["Scientific name"]+" ("+df['Common name']+")"
df["Code"] = "Dutch Buffalo Farm"
df["Genetics"] = "Species"
df["Terms"] = "Species" # think this shoulda been lowercase

originCodeCondition = [df["Origin code"] == "1",
    df["Origin code"] == "2",
    df["Origin code"] == "3",
    df["Origin code"] == "4",
    df["Origin code"] == "5"]
originCodeChoice = ["NC - Piedmont","NC - Piedmont","US - Southeast","US - SE - Nativar","Exotic"]
df["Origin"] = np.select(originCodeCondition,originCodeChoice,"")
df["Use"] = "naturalizing"
df["Supplier notes"] = "Dutch Buffalo Farm list 2022, with origin code "+df["Origin code"]
```

```python
# Get copy-past view for Perennial airtable -- then do manual merge on species, making sure to keep name the same
colList = ['Name', 'Code','Genetics','Scientific name','Terms','Origin',"Use","Supplier notes"]
df[colList].set_index("Name").head(50)
```
